```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("adult_with_headers (1).csv")
```

```python
df.replace('?', np.nan, inplace=True)
df.isnull().sum()
```

```python
numb_cols = df.select_dtypes(include=['int64', 'float64']).columns
cato_cols = df.select_dtypes(include=['object']).columns

for col in numb_cols:
    df[col].fillna(df[col].median(), inplace=True)

for col in cato_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

num_features = df[numb_cols]
```

```python
#Standard Scaling   
std_scaler = StandardScaler()
df_standard_scaled = pd.DataFrame(
    std_scaler.fit_transform(num_features),
    columns=numb_cols
)
```

```python
#Min-Max Scaling
minmax_scaler = MinMaxScaler()
df_minmax_scaled = pd.DataFrame(
    minmax_scaler.fit_transform(num_features),
    columns=numb_cols
)
```

```python
df[cato_cols].nunique()
```

```python
#One-Hot Encoding to categorical variables with less than 5 categories.
low_cardinality_cols = [col for col in cato_cols if df[col].nunique() <= 5]

df1 = pd.get_dummies(df, columns=low_cardinality_cols, drop_first=True)
```

```python
#Label Encoding for categorical variables
from sklearn.preprocessing import LabelEncoder

high_cardinality_cols = [col for col in cato_cols if df[col].nunique() > 5]

le = LabelEncoder()
for col in high_cardinality_cols:
    df1[col] = le.fit_transform(df1[col])
```

```python
df1['education_per_year'] = df1['education_num'] / df1['age']
```

```python
df1['work_intensity'] = df1['hours_per_week'] * df1['age']
```
